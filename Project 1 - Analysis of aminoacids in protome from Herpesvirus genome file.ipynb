### Parse 3 herpesvirus_genome.json file and find the frequency of each amino acid in the proteome of the herpesvirus

```python
import json

# opening .json file 
f = open(r'...\3 herpesvirus_genome.json', 'r')
data = json.load(f)
f.close()
```

```python
# data is stored in a dictionary
data
```

```python
# extracting data under key 'coding_region'
coding_regions = data['coding_regions']
coding_region
```

```python
# appending translation sequences from coding region into a list 
translations = []
for i in range(0,len(coding_regions)):
    coding_region = coding_regions[i]
    for key, value in coding_region.items():
        if key == "translation":
            translations.append(value)
translations
```

```python
# dictionary with amino acids as keys and values equal to 0
aminoacids_dict ={
'A': 0, 'C': 0, 'D': 0, 'E': 0, 'F': 0,
'G': 0, 'H': 0, 'I': 0, 'K': 0, 'L': 0, 
'M': 0, 'N': 0, 'P': 0, 'Q': 0, 'R': 0, 
'S': 0, 'T': 0, 'V': 0, 'W': 0, 'Y': 0,
}
 
# function used to count occurences for each aminoacid, takes two parameters dictionary of amino acids and list of transaltions
def count_freq(a_dict,t_list):
    for i in t_list:
        for aa in i:
            a_dict[aa] += 1
            
            
count_acids = aminoacids_dict.copy()
count_freq(count_acids, translations)
                   
count_acids
# sum(count_acids.values())
```

```python
# calculate frequences for each AA in translation sequence
def get_precentage_dict(a_dict):
    total_sum = sum(a_dict.values())
    new_dict = a_dict.copy()
    for key, value in a_dict.items():
        new_dict[key] = round((value/total_sum)*100,2)
    return new_dict
    
precentage_amino_total = get_precentage_dict(count_acids)
precentage_amino_total
```

### Find the frequency of each amino acid, but separate the results based on the three categories of proteins: envelope, membrane, capsid.

```python
# create three lists for each category 
envelope = []
membrane = []
capsid = []

# appending translation sequences to appropriate list, based on the category (which is specifed as a first word in value of the "product" key)
for i in range(0,len(coding_regions)):
    coding_region = coding_regions[i]
    for key, value in coding_region.items():
        if coding_region["product"].split()[0] == "envelope":
            envelope.append(coding_region["translation"])
        elif coding_region["product"].split()[0] == "membrane":
            membrane.append(coding_region["translation"])
        elif coding_region["product"].split()[0] == "capsid":
            capsid.append(coding_region["translation"])   

# number of translation sequences in each category
print(len(envelope))
print(len(membrane))
print(len(capsid))
```

**Calculatation of AA frequencies for each category (envelope, capsid and membrane)**

```python
envelope_freq = aminoacids_dict.copy()
count_freq(envelope_freq, envelope)
precent_envelope = get_precentage_dict(envelope_freq)
precent_envelope
```

```python
membrane_freq = aminoacids_dict.copy()
count_freq(membrane_freq, membrane)       
precent_membrane = get_precentage_dict(membrane_freq)
precent_membrane
```

```python
capsid_freq = aminoacids_dict.copy()
count_freq(capsid_freq, capsid)       
precent_capsid = get_precentage_dict(capsid_freq)     
precent_capsid
```

### What are amino acids with the highest and the lowest frequencies in all four categories? (overall, envelope, membrane, capsid)


Compare the polarity of these different amino acids

Link: https://www.imgt.org/IMGTeducation/Aide-memoire/_UK/aminoacids/IMGTclasses.html#:~:text=There%20are%202%20IMGT%20'Polarity,%2C%20P%2C%20W%2C%20V.

```python
# two lists containing polar and nonpolar AA
polar = ["R", "N", "D", "Q", "E", "H", "K", "S", "T", "Y"]
nonpolar = ["A","C", "G", "I", "L", "M", "F", "P", "W", "V"]

# creating a dictionary of AAs having with value polar or nonpolar
polarity_dict = aminoacids_dict.copy()
for i in polarity_dict:
    if i in polar:
        polarity_dict[i] = "polar"
    else:
        polarity_dict[i] = "nonpolar"
        
# function to check whether specific AA 
def check_polarity(x):
    if x in polar:
         return "polar"
    else:
        return "nonpolar"
    
# count number of polar and nonpolar AAs from dictionary containing occurences for each AA
def count_polarity(a_dict):
    count_polar = 0
    count_nonpolar = 0
    for i in a_dict:
        if i in polar:
            count_polar += a_dict[i]
        else:
            count_nonpolar += a_dict[i]
            
    print("Total number of polar aminoacids:", count_polar, 
          "\nTotal number of nonpolar aminoacids:", count_nonpolar)
    
# results
count_polarity(count_acids)
```

```python
# function to get highest and lowest frequency of AA in a dictionary
def get_min_max_freq(a_dict):
    print(" Highest frequency\n  Amino acid:",
      max(a_dict, key=a_dict.get), "\n  Frequency:", max(a_dict.values()), "\n  Precentage:" 
          , max(get_precentage_dict(a_dict).values()), "%\n  Polarity:" ,check_polarity(max(a_dict, key=a_dict.get)),
     "\n Lowest frequency:\n  Amino acid:",
      min(a_dict, key=a_dict.get), "\n  Frequency:", min(a_dict.values()), "\n  Precentage:" 
          , min(get_precentage_dict(a_dict).values()), "%", "%\n  Polarity:" ,check_polarity(min(a_dict, key=a_dict.get)), "\n")

print("Overall category")
get_min_max_freq(count_acids)
print("Envelope category")
get_min_max_freq(envelope_freq)
print("Membrane category")
get_min_max_freq(membrane_freq)
print("Capsid category")
get_min_max_freq(capsid_freq)
```

Most and least frequent aminoacids for each group are A (Alanine) and W (Tryptophan). A and W aminoacids are nonpolar.

### What is the average hydrophobicity of four different protein groups? (overall, envelope, membrane, capsid)

Calculate the average hydrophobicity per every group  add all hydrophobicity measures for all 
amino acids for all proteins per group, and take the average for the entire group. Comment on the 
comparison.
Use the following hydrophobicity scale: (I=4.5, V=4.2, L=3.8, F=2.8, C=2.5, M=1.9, A=1.8, G=-0.4, T=-
0.7, S=-0.8, W=-0.9, Y=-1.3, P=-1.6, H=-3.2, E=-3.5, Q=-3.5, D=-3.5, N=-3.5, K=-3.9, R=-4.5)


```python
# dictionary containing a hydrophobicity value of each AA
hydrophopic_dict = {"I" : 4.5, "V": 4.2, "L" : 3.8, "F":2.8, "C":2.5, "M":1.9, "A":1.8, "G":-0.4, 
                     "T":-0.7, "S":-0.8, "W":-0.9, "Y":-1.3, "P":-1.6, "H":-3.2, "E":-3.5, "Q":-3.5, 
                     "D":-3.5, "N":-3.5, "K":-3.9, "R":-4.5}

# function to calculate a average hydrophobicity using a dictionary that contains 
def get_avg_hydrophobic(aa_dict):
    x = sum(aa_dict.values())
    for key, value in aa_dict.items():
        aa_dict[key] = hydrophopic_dict[key] * value
    print(sum(aa_dict.values())/x) 

print("Overall:")
get_avg_hydrophobic(count_acids.copy()) 
print("Memebrane")
get_avg_hydrophobic(membrane_freq.copy())
print("Envelope:")
get_avg_hydrophobic(envelope_freq.copy())
print("Capsid")
get_avg_hydrophobic(capsid_freq.copy())   
```

Based on the result for hydrophobicity of four categories (overall, membrane, envelope and capsid), we can see that envelope is the most hydrophobic category, while capsid category is the most hydrophilic category.

Herpesvirus is a enveloped virus, it has an envelope or lipid bilayer. Lipid bilayer is made of phosolipids, head of phospholipid is a hydrophobic while tails are hydrophilic making the outer parts of the layer hydrophobic while inner layer is hydrophilic (but water-free). 
Envelope proteins are found on the lipid bilayer (depending on the type of a protein they can be found also inside of the lipid bilayer like lipidbound proteins but this type of the proteins are very rare, beacuse proteins are there to interact with outside enviroment, but this type can not interact with outside nor with the inside of the cell beacuse of the structure of a lipid bilayer), they allows virus to infect cells, also envelope proteins are used in intracellular transport (like channel or carrier type of proteins). Envelope proteins are mainly specific for each hypervirus type. Also, herpesvirus envelopes are estimated to carry between 12-20 membrane proteins, which we can see from our results are the second the most hydrophobic proteins. So, herpesvirus has outside layer which is a hydrophobic, so just washing hands with a water is not enough to destroy a virus, but also making this virus not transmittable in water.

Envelope of a protein is protecting inner part of a virus which is called nucleocapsid (and tegument which is a part between nucleocapsid and envelope). Capsid is a protein coat, protecting a genomic material of a virus (in the case of Herpesvirus is a DNA) and creating a nucleocapsid of a virus. Viral capsid is in general water-soluble, and that is a reason why this category has lowest hydrophobicity than other categories.
